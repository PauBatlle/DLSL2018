```python
tsid = train.sentence_id.tolist()
tbe = train.before.tolist()
taf = train.after.tolist()
max_id = tsid[-1]
div = int(6*max_id/7)
```

```python
lsb = []
testb = []
for i in range(div):
    lsb.append([])
for i in range(div, max_id+1):
    testb.append([])
```

```python
clase = train["class"]
```

```python
encoder = LabelEncoder()
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(clase)
encoded_Y = encoder.transform(clase)
# convert integers to dummy variables (i.e. one hot encoded)
labels = np_utils.to_categorical(encoded_Y)
```

```python
class1 = []
class2 = []
vocab = []
for i in range(len(tsid)):
    sid = tsid[i]
    vocab.append(str(tbe[i]))
    if (sid < div):
        lsb[sid].append(str(tbe[i]))
        class1.append(labels[i])
    else:
        testb[sid-div].append(str(tbe[i]))
        class2.append(labels[2])
```
