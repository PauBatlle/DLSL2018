```python
import pandas as pd
import numpy as np
import gensim
import logging
from tqdm import tqdm as t
```

```python
train = pd.read_csv('en_train.csv')
output = pd.read_csv('output.csv')
test = pd.read_csv('en_test.csv')
```

```python
sentences_id = train.sentence_id.tolist()
sentences_before = train.before.tolist()
sentences_after = train.after.tolist()
max_id = len(sentences_id)
```

```python
text_before = []
text_after = []
for i in range(max_id):
    text_before.append([])
    text_after.append([])
```

```python
for i in range(len(train)):
    sid = sentences_id[i]
    text_before[sid].append(str(sentences_before[i]))
    text_after[sid].append(str(sentences_after[i]))
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = gensim.models.Word2Vec(text_before, size=500, window=5, min_count=3, workers=4)
model2 = gensim.models.Word2Vec(text_after, size=500, window=5, min_count=3, workers=4)
model.save('trad.1.1')
model2.save('trad.1.2')
```

```python
total = 0
fallos = 0
for i in t(range(10000)):
    word = test.before[i]
    if word in model.wv.vocab:
        v = (model.wv[word])
        trad = model2.wv.similar_by_vector(v)[0][0]
    else:
        trad = word
    total = total+1
    if trad != output.after[i]:
        fallos = fallos + 1
print("Accurancy: " + str((1-fallos/total)*100)+ " %")
```
